##  <span style="color:blue"> Exercise 2.1

Integral to evaluate:
$$I = \int_0^1 \frac{\pi}{2}\cos(\frac{\pi}{2} x) dx = 1$$

This task has been carried out, as requested

1) using a U[0,1] PDF
$$f(x)=\frac{\pi}{2}\cos(\frac{\pi}{2} x), \hspace{1cm} p(x) = 1$$
2) via importance sampling
$$f'(x)=\frac{f(x)p(x)}{p'(x)}, \hspace{1cm} p'(x)= \frac{100}{51}\left ( \frac{101}{100}-x \right )$$

The idea behind this latter choice is sketched as follows.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.arange(0.0, 1.01, 0.003)
f = (np.pi/2)*np.cos(np.pi * x/2)
g = (100/51)*((101/100)-x)

plt.plot(x,f,label="f(x)")
plt.plot(x,g,label="p'(x)")
plt.xlabel('x')
plt.grid(True)
plt.legend()

plt.show()
```

- The goal of importance sampling is to choose a PDF which resembles the original function to integrate. A quick look at the figure above suggests that a natural choice is the straight line connecting $(0,\frac{\pi}{2})$ and $(1,0)$ (to be correctly normalized *a posteriori*).
- But then a problem arises: such PDF equals 0 inside the domain of integration (more precisely, at the right border). This cannot be accepted, since in importance sampling the function chosen as PDF also appears at the denominator of the new function $f'(x)$ to be sampled, resulting in an untreatable singularity.
- Nonetheless, this problem can be circumvented by letting our trial PDF intersect the x axis slightly beyond $x=1$, for example in $x=1+\frac{1}{100}=\frac{101}{100}$. After normalization, our PDF takes the form above.

- One might now ask whether this game is worth the candle: that is, if the error $\sigma_I$ is reduced and to what extent. It can be easily shown that 
$$\sigma_{UNIF}^2 = \int_0^1 f^2(x)p(x) - I^2 \simeq 0.2337$$
$$\sigma_{IS}^2 = \int_0^1 \left ( \frac{f(x)p(x)}{p'(x)} \right )^2 p'(x) - I^2 \simeq 0.01438$$

  so that the ratio for equal number of throws is $\Delta = \sqrt{ \frac{\sigma_{IS}^2}{\sigma_{UNIF}^2}} \simeq 0.248$: **the error is reduced by a factor 4. <span style="color:red"> REALLY? check figure! <span>**
    

- Lastly, in order to sample with $p'(x)$ one has to exploit the cumulative inversion method. 
$$F_x(x)=\int_{-\infty}^x p'(x')dx' = \frac{1}{51} \left ( 101x - 50x^2\right ) := y \in \left [0,1 \right ]$$ 
 so that $$x=\frac{1}{100}\left ( 101 \pm \sqrt{10201-10200y}\right )$$

  Here, the right sign to choose is the $-$ sign, the only one that guarantees $x \in \left [0,1 \right ] $.

```python
N = 100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
sum_prog2 = np.zeros(N)
err_prog2 = np.zeros(N)
steps = np.arange(N)

file = open("./LSN_02.1/output_1.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split("   ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
file.close()
    
file2 = open("./LSN_02.1/output_2.txt")
line2 = file2.readlines()
for i in range(N):
    b = line2[i].split("   ")
    sum_prog2[i] = b[1]
    err_prog2[i] = b[2]
file2.close()
    
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.xlabel("N° blocks")
plt.ylabel("I-1")
plt.ylim(-0.0145, 0.006)
plt.grid(True)
plt.title("Uniform PDF")

plt.subplot(1, 2, 2)
plt.errorbar(steps, sum_prog2, yerr = err_prog2, color='orange')
plt.xlabel("N° blocks")
plt.ylabel("I-1")
plt.ylim(-0.0145, 0.006)
plt.grid(True)
plt.title("Importance sampling")

plt.subplots_adjust(wspace=0.25)
plt.show()
```

##  <span style="color:blue"> Exercise 2.2

```python
from scipy.optimize import curve_fit

def f(x,k):
    return k*(x**0.5)

file = open("./LSN_02.2/output_2.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split("   ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
file.close()
    
file2 = open("./LSN_02.2/output_2.txt")
line2 = file2.readlines()
for i in range(N):
    b = line2[i].split("   ")
    sum_prog2[i] = b[1]
    err_prog2[i] = b[2]
file2.close()

plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.errorbar(steps, sum_prog, yerr = err_prog, label='simulation')
plt.xlabel("N° steps")
plt.ylabel("$sqrt(<(|r_N|^2)>$")
plt.grid(True)
plt.title("Discrete cubic lattice")
p_opt,_ = curve_fit(f, x, sum_prog)
plt.plot(x,f(x,p_opt[0]),label='fit k*sqrt(N)', color='red')
plt.legend()

plt.subplot(1, 2, 2)
plt.errorbar(steps, sum_prog2, yerr = err_prog2, label='simulation', color='orange')
plt.xlabel("N° steps")
plt.ylabel("$sqrt(<(|r_N|^2)>$")
plt.grid(True)
plt.title("Continuum")
p_opt2,_ = curve_fit(f, x, sum_prog)
plt.plot(x,f(x,p_opt2[0]),label='fit k*sqrt(N)', color='red')
plt.legend()

plt.subplots_adjust(wspace=0.25)
plt.show()

print("FIT PARAMETER k")
print("Discrete cubic lattice:")
print(p_opt[0])
print("Continuum:")
print(p_opt[0])
```

Here, the mean square distance of a random walker has been worked out and fitted with the expected behaviour $\propto k\sqrt{N}$.

We observe perfect agreement, which does suggest diffusive behaviour.
