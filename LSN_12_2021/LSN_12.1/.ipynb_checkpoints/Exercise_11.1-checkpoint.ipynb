```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect

import numpy as np

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.0 # noise standard deviation
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)

# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=300,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r',label="predicted")
plt.plot(x_valid, y_target,label="valid")
plt.grid(True)
plt.legend()
plt.show()
```

### Osservazioni

- Si pu vedere facilmente che, all'aumentare del rumore $\sigma$ del training set, i valori di *training loss* e *validation loss* aumentano notevolmente e aumenta anche il gap presente tra i due valori. Pi rimane basso tale valore e pi si ha un errore basso nel *training* e nel *test set*.


- Tenendo fissato il valore di $N_{train}$ e aumentando quello di *epochs* si vede che all'aumentare delle *epochs* i due valori di perdita rimangono pressoch costanti.


- Si pu vedere che con un numero di $N_{epochs}$ dell'ordine delle centinaia, per esempio 300, e un numero di $N_{train}$ uguale, per esempio a 10, si ha che la funzione del *training set* converge sempre di pi ad un valore, mentre quella del *test set* diverge dopo aver raggiunto un punto stazionario di minimo. Questo suggerisce il fatto che si sia raggiunta una condizione di overfitting. Inoltre mantenere un numero di $N_{train}$ di questo ordine di grandezza impedisce di ottenere una misura accurata, come  possibile vedere dalla test accuracy.

Riassumendo si consiglia di mantenere un valore $N_{train}$ sufficiente a non causare la divergenza della *loss function* e un valore $\sigma$ sufficientemente basso per mantenere il valore dell'errore sul *training set* inferiore. All'aumentare del valore di $N_{epochs}$ le due funzioni convergono sempre di pi ad un valore di equilibrio.
