```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x,y):
    return math.sin(x**2+y**2)

# generate training inputs
N_t=1000
N_v=10
np.random.seed(0)
x_train = np.random.uniform(-1.5, 1.5, N_t)
y_train = np.random.uniform(-1.5, 1.5, N_t)
x_valid = np.random.uniform(-1.5, 1.5, N_v)
y_valid = np.random.uniform(-1.5, 1.5, N_v)

xy_valid=np.array([
    [x_valid[i],y_valid[i]] for i in range(N_v)
])
xy_train=np.array([
    [x_train[i],y_train[i]] for i in range(N_t)
])

z_target = ([
    f(x_valid[i],y_valid[i]) # ideal (target) function
    for i in range(N_v)
])

sigma = 0.5 # noise standard deviation
# actual measures from which we want to guess regression parameters
z_train = np.array([
    np.random.normal(f(x_train[i],y_train[i]), sigma)
    for i in range(N_t)
]) 
z_valid = np.array([
    np.random.normal(f(x_valid[i],y_valid[i]), sigma)
    for i in range(N_v)
])

#x_true,y_true and z_true to be used for the 3Dplot comparison
x_true = np.random.uniform(-1.5, 1.5, N_t)
y_true = np.random.uniform(-1.5, 1.5, N_t)

z_true=([
    f(x_true[i],y_true[i]) for i in range(N_t)
])

# plot target dataset
ax = Axes3D(plt.figure())
ax.plot_trisurf(x_true, y_true, z_true,alpha=0.3,color='g')
ax.scatter(x_valid, y_valid, z_valid, color='r')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("$f(x,y)=sin(x^2+y^2)$")
ax.view_init(35, 35)
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(100, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(1, input_shape=(2,)))
model.add(Activation("relu"))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=xy_train, y=z_train,
          batch_size=128, epochs=500,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, z_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(xy_valid, z_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# evaluate model with the exact curve
score = model.evaluate(xy_valid, z_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = np.random.uniform(-1.5, 1.5, 1000)
xy_predicted = np.array([
    [x_predicted[i],y_predicted[i]] for i in range(1000)
])
z_predicted = model.predict(xy_predicted)

ax = Axes3D(plt.figure())
ax.plot_trisurf(x_true, y_true, z_true,alpha=0.3,label="true")
ax.scatter(x_predicted, y_predicted, z_predicted, color='r',label="predicted")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Predicted results")
#ax.legend()
ax.view_init(15, 35)
```

### Osservazioni


- Si osserva che è fondamentale utilizzare una deviazione standard dalla media $\sigma$ non nulla all'interno del *training set*. Nell'esempio riportato sopra è stato utilizzato un valore pari a $0.5$, il che ha permesso al modello di dedurre in modo più efficace la curva che si voleva descrivere. Analoghi tentativi condotti con valori di $\sigma=0$ e $\sigma=0.05$ hanno mostrato una pessima previsione del risultato. 


- Naturalmente è necessario un modello sufficientemente complesso, ovvero tale da dedurre il comportamento di una funzione altrimenti non riconoscibile. Di conseguenza è essenziale utilizzare un numero di layers e/o neuroni adeguati.


- Aumentare il numero di $N_{epochs}$ consente di migliorare l'errore relativo al *validation set*. Come al solito occorre stare attenti al rischio di overfitting.
