### Autocorrelation function

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

def autocorrelation(vet,tau):
    ntot=len(vet)
    nstop=ntot-tau
    ave=0
    ave2=0
    ave_t=0
    ave_t2=0
    ave_mix=0

    for i in range(ntot):
        ave+=vet[i]
        ave2+=vet[i]**2
        if(i+tau<ntot):
            #ave_t+=vet[i+tau]
            #ave_t2+=vet[i+tau]**2
            ave_mix+=vet[i]*vet[i+tau]
    ave/=ntot
    ave2/=ntot
    
    #ave_t/=nstop
    #ave_t2/=nstop
    ave_mix/=nstop
    
    ac=(ave_mix-ave**2)/(ave2-ave**2)
    
    return ac
```

## Autocorrelation of $U/N$ and $P$
### for solid, liquid and gas phase

```python
N=100000
pressure=np.zeros(N)
int_ene=np.zeros(N)

def exp_fit(x,k):
    return k*math.exp(-x)

n_tau=100
ac_ene=np.zeros(n_tau)
ac_press=np.zeros(n_tau)
x=np.arange(n_tau)

for phase in range(1,4):
    if(phase==1):
        file=open("InstantData.gas")
    if(phase==2):
        file=open("InstantData.liquid")
    if(phase==3):
        file=open("InstantData.solid") 
        
    line=file.readlines()
    for i in range(N):
        data=line[i].split()
        int_ene[i]=data[0]
        pressure[i]=data[1] 
    
    for t in range(n_tau):
        ac_ene[t]=autocorrelation(int_ene,t)
        ac_press[t]=autocorrelation(pressure,t)
    
    fig, (ene, press) = plt.subplots(1, 2, figsize=(19, 5))
    
    ene.plot(ac_ene,label='Simulation')
    ene.set_xlabel("t")
    ene.set_ylabel("Ac(t)")
        
    #p_opt, _ = curve_fit(exp_fit, x, ac_ene)
    #ene.plot(x,exp_fit(x,p_opt[0]),label='Fit')
    #ene.legend()
        
    press.plot(ac_press,label='Simulation')
    press.set_xlabel("t")
    press.set_ylabel("Ac(t)")
        
    #p_opt,_ = curve_fit(exp_fit, x, ac_press)
    #press.plot(x,exp_fit(x,p_opt[0]),label='Fit')
    #press.legend()
    
    if phase==1:
        ene.set_title('Internal Energy autocorrelation \n Gas phase')
        press.set_title('Pressure autocorrelation \n Gas phase')
    if phase==2:
        ene.set_title('Internal Energy autocorrelation \n Liquid phase')
        press.set_title('Pressure autocorrelation \n Liquid phase')
    if phase==3:
        ene.set_title('Internal Energy autocorrelation \n Solid phase')
        press.set_title('Pressure autocorrelation \n Solid phase')
    
file.close()
```

## Statistical uncertainties of $U/N$ and $P$

```python
from array import array

def error(AV,AV2,n):  
    if n==0:
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n)

L=[]   #Lengths of the blocks
n=[]   #Blocks
for i in range(2,N):
    if N%i==0:
        n.append(i)
        L.append(int(N/i))   #By this way we select all the integer lengths 
L=L[::-1]   #Not essential, but only to have lengths increasing
n=n[::-1]

ave_ene = np.zeros(n[0])   #We make this array of the biggest size to avoid using a dynamical array
ave_press = np.zeros(n[0])

err_ene=np.zeros(len(L))
err_press=np.zeros(len(L))

for phase in range(1,4):
    if(phase==1):
        file=open("InstantData.gas")
    if(phase==2):
        file=open("InstantData.liquid")
    if(phase==3):
        file=open("InstantData.solid")

    line=file.readlines()
    for i in range(N):
        data=line[i].split()
        int_ene[i]=data[0]
        pressure[i]=data[1]
    
    for l in range(len(L)):
        
        for i in range(n[0]):
            ave_ene[i]=0
            ave_press[i]=0
        
        for i in range(n[l]):
            for j in range(L[l]):
                k = j+i*L[l]
                ave_ene[i] += int_ene[k]
                ave_press[i] += pressure[k]
            ave_ene[i] /= L[l]  
            ave_press[i] /= L[l]
        
        sum_p =0
        su2_p =0
        
        for i in range(n[l]):
            sum_p += ave_ene[i]
            su2_p += ave_ene[i]**2 
        sum_p /= n[l]
        su2_p /= n[l]
        err_ene[l]=math.sqrt((su2_p - sum_p**2)/n[l])   #Statistical uncertainty on energy
    
        sum_p=0
        su2_p=0
    
        for i in range(n[l]):
            sum_p += ave_press[i]
            su2_p += ave_press[i]**2 
        sum_p /= n[l]
        su2_p /= n[l]
        err_press[l]=math.sqrt((su2_p - sum_p**2)/n[l])   #Statistical uncertainty on pressure

                 
    fig, (pl_ene, pl_press) = plt.subplots(1, 2, figsize=(19, 5))

    pl_ene.scatter(L,err_ene)
    pl_ene.set_xlabel("L")
    pl_ene.set_ylabel('uncertainty')
                 
    pl_press.scatter(L,err_press)
    pl_press.set_xlabel("L")
    pl_press.set_ylabel('uncertainty')
    
    if phase==1:
        pl_ene.set_title('Statistical uncertainty on Internal Energy \n Gas phase')
        pl_press.set_title('Statistical uncertainty on pressure \n Gas phase')
        pl_ene.axes.set_ylim([0,0.001])
        pl_press.axes.set_ylim([0,0.0002])
    if phase==2:
        pl_ene.set_title('Statistical uncertainty on Internal Energy \n Liquid phase')
        pl_press.set_title('Statistical uncertainty on pressure \n Liquid phase')
        pl_ene.axes.set_ylim([0,0.003])
        pl_press.axes.set_ylim([0,0.015])
    if phase==3:
        pl_ene.set_title('Statistical uncertainty on Internal Energy \n Solid phase')
        pl_press.set_title('Statistical uncertainty on pressure \n Solid phase')   
        pl_ene.axes.set_ylim([0,0.002])
        pl_press.axes.set_ylim([0,0.008])
```

We observe that the uncertainty remains approximately constant with the increasing lengths of the blocks.

## Averages values of $U/N$, $P$ and $g(r)$

We begin plotting the graphs of $g(r)$ in the three phases, later we plot the graphs of internal energy $U/N$ and pressure $P$ for Argon and Krypton.

### $g(r)$

```python
nbins=100
ave_g=np.zeros(nbins)
err_g=np.zeros(nbins)
r=np.zeros(nbins)

for phase in range(1,4):
    file=open("output.gave.0")
    line=file.readlines()
    
    for i in range(nbins):
        data=line[nbins*(phase-1)+i].split()
        r[i]=data[0]
        ave_g[i]=data[1]
        err_g[i]=data[2]
    
    plt.figure()    
    
    if phase==1:
        plt.title("Radial distribution probability $g(r)$ \n Gas phase")
        plt.errorbar(r,ave_g,yerr=err_g)
    if phase==2:
        plt.title("Radial distribution probability $g(r)$ \n Liquid phase")
        plt.errorbar(r,ave_g,yerr=err_g, color='r')
    if phase==3:
        plt.title("Radial distribution probability $g(r)$ \n Solid phase")
        plt.errorbar(r,ave_g,yerr=err_g, color='g')
    
    plt.xlabel("r/$\sigma$")
    plt.ylabel('g(r)')
```

### Argon: $U/N$, $P$

```python
K=1.3806*10**(-23)   #Boltzmann constant

#Data about Argon
sigma=0.34*10**(-9)
rap=120
mass=39.948*1.66054*10**(-27)
eps=rap*K

nblocks=100

x=np.arange(nblocks)
ave_ene=np.zeros(nblocks)
err_ene=np.zeros(nblocks)

ave_press=np.zeros(nblocks)
err_press=np.zeros(nblocks)

for phase in range(1,4):
    
    file1=open("output.epot.0")
    file2=open("output.pres.0")
    
    line1=file1.readlines()
    line2=file2.readlines()
    
    for i in range(nblocks):
        data1=line1[nblocks*(phase-1)+i].split()
        data2=line2[nblocks*(phase-1)+i].split()
        ave_ene[i]=data1[2]
        err_ene[i]=data1[3]
        ave_press[i]=data2[2]
        err_press[i]=data2[3]
    
    fig, (fig_ene, fig_press) = plt.subplots(1, 2, figsize=(15, 4))
    
    if phase==1:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps)
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3))
        fig_ene.set_title('Internal energy U/N \n Gas phase')
        fig_press.set_title('Pressure P \n Gas phase')
    if phase==2:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps, color='r')
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3), color='r')
        fig_ene.set_title('Internal energy U/N \n Liquid phase')
        fig_press.set_title('Pressure P \n Liquid phase')
    if phase==3:
        fig_ene.errorbar(x,ave_ene*eps,yerr=err_ene*eps,color='g')
        fig_press.errorbar(x,ave_press*eps/(sigma**3),yerr=err_press*eps/(sigma**3),color='g')
        fig_ene.set_title('Internal energy U/N \n Solid phase')
        fig_press.set_title('Pressure P \n Solid phase')
    
    fig_ene.set_xlabel("blocks")
    fig_ene.set_ylabel("U/N (J)")
    
    fig_press.set_xlabel("blocks")
    fig_press.set_ylabel('P (Pa)')
        
```
