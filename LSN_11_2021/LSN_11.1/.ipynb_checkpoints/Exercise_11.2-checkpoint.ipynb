```python
import os
os.environ['KMP_DUPLICATE_LIB_OK']='True'

import numpy as np

# target parameters of f(x) = a*x^3 + b*x^2 + c*x + d
a=3
b=-2
c=-3
d=4

def f(x):
    return a*x**3 + b*x**2 + c*x + d

# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 1000)
x_valid = np.random.uniform(-1, 1, 10)
x_valid.sort()
y_target = f(x_valid) # ideal (target) function

sigma = 0.0 # noise standard deviation
y_train = np.random.normal(f(x_train), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(f(x_valid), sigma)

# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

## Comparison of models with different number of neurons and layers

In the following code we create 5 different models and we predict the results inside and outside the fitting range.
The models are:

(1) 0 hidden layers (1 input and 1 output layer)

(2) 1 hidden layer with 5 neurons

(3) 1 hidden layer with 50 neurons

(4) 2 hidden layers with 5 and 2 neurons respectively

(5) 3 hidden layers with 5, 2 and 2 neurons respectively

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

models=[]
units=[1,50,100]

#The following models have different number of neurons in the only hidden layer
for i in range(5):
    #The second and third have an increasing number of neurons, but one fixed hidden layer
    if i<3:
        model= tf.keras.Sequential()
        model.add(Dense(units[i], input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(1, input_shape=(1,)))
        model.add(Activation("elu"))
    #The fourth and the fifth have more (hidden) layers
    if i==3:
        model= tf.keras.Sequential()
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(50, input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(1, input_shape=(1,)))
        model.add(Activation("elu"))
    if i==4:
        model= tf.keras.Sequential()
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation("elu"))
        model.add(Dense(1, input_shape=(1,)))
        model.add(Activation("elu"))
        
    models.append(model)

# compile the models choosing optimizer, loss and metrics objects
for model in models:
    model.compile(optimizer='sgd', loss='mse', metrics=['mse'])


# get a summary of our composed model
i=1
for model in models:
    print("Model ",i)
    model.summary()
    print()
    i+=1
```

```python
# fit the models using training dataset
# report training progress against validation data
history=[]
for model in models:
    history.append(model.fit(x=x_train, y=y_train, 
              batch_size=64, epochs=200,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid)))
```

```python
# return weights
for model in models:
    model.get_weights()
```

```python
i=1
for model in models:
    print("Model ",i)
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

    # print performance
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print()
    i+=1
```

```python
i=1
for model in models:
    print("Model ",i)
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

    # print performance
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print()
    i+=1
```

The number near the names contained in the legends refer to the model used.

```python
# look into training history
# Plot training & validation loss values

fig, (fig_1, fig_2) = plt.subplots(1, 2, figsize=(18,6))

i=1
plt.figure(figsize=(12,9))
for h in history:
    if i<=3:
        fig_1.plot(h.history['loss'],label="loss "+str(i))
        fig_1.plot(h.history['val_loss'],label="val_loss "+str(i))
    else:
        fig_2.plot(h.history['loss'],label="loss "+str(i))
        fig_2.plot(h.history['val_loss'],label="val_loss "+str(i))
    i+=1

fig_1.legend(loc='best')
fig_1.set_title("Model loss \n Models 1,2,3")
fig_1.set_ylabel('Loss')
fig_1.set_xlabel('Epoch')

fig_2.legend(loc='best')
fig_2.set_title("Model loss \n Models 4,5")
fig_2.set_ylabel('Loss')
fig_2.set_xlabel('Epoch')
```

```python
x_predicted_1 = np.random.uniform(-1, 1, 100)
x_predicted_2 = np.random.uniform(1, 3, 100)

x_valid_2=np.random.uniform(1,3,10)
x_valid_2.sort()
y_target_2=f(x_valid_2)

fig, (fig_1, fig_2) = plt.subplots(1, 2, figsize=(18,6))

i=1
for model in models:
    y_predicted_1 = model.predict(x_predicted_1)
    y_predicted_2 = model.predict(x_predicted_2)
    fig_1.scatter(x_predicted_1, y_predicted_1,label="Model "+str(i))
    fig_2.scatter(x_predicted_2, y_predicted_2,label="Model "+str(i))
    i+=1

fig_1.plot(x_valid, y_target,label="target",color='y')
fig_1.set_title("Predicted results \n Range $[-1,1]$")
fig_1.legend()


fig_2.plot(x_valid_2, y_target_2,label="target",color='y')
fig_2.set_title("Predicted results \n Range $[1,3]$")
fig_2.legend()
```

### Osservazioni

- Si pu vedere che per modelli con numero di layers e/o neuroni maggiore le nostre predizioni risultano migliori nel range di studio, $[-1,1]$. Con un alto numero di parametri in gioco si riesce a migliorare l'accuratezza della nostra previsione, come  possibile vedere dai risultati ottenuti per il *Modello 4* e il *Modello 5*. Infatti, dal grafico della *validation loss* relativo al primo modello si vede come questo sia completamente inefficiente. Nei modelli successivi si vede come sia possibile ottenere un miglioramento sul *validation set*.


- Si nota che la differenza nella predizione tra il *Modello 4* e il *Modello 5* risulta minima. Si deduce che al di sopra di un certo livello di complessit i risultati non cambiano sensibilmente.


- Si pu vedere che tutti i modelli usati sono inefficienti al di fuori del fitting range. 

## Comparison of models with different activation functions, optimizers and loss function

In the following code we create 7 different models and we predict the results inside and outside the fitting range. All models below share the same number of layers and neurons for each layer.
The models have:

(1) {activation_function='elu', optimizer='sgd', loss='mse'}

(2) {activation_function='tanh', optimizer='sgd', loss='mse'}

(3) {activation_function='softmax', optimizer='sgd', loss='mse'}

(4) {activation_function='elu', optimizer='adam', loss='mse'}

(5) {activation_function='elu', optimizer='RMSprop', loss='mse'}

(6) {activation_function='elu', optimizer='sgd', loss='mape'}

(7) {activation_function='elu', optimizer='sgd', loss='msle'}

```python
models=[]

act_func=["elu","tanh","softmax"]
opt=["sgd","adam","RMSprop"]
loss=["mse","mape","msle"]  #categorichal cross entropy

for i in range(7):
    if i<3:
        model= tf.keras.Sequential()
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[i]))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[i]))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[i]))
        model.add(Dense(1, input_shape=(1,)))
        model.add(Activation(act_func[i]))
    else:
        model= tf.keras.Sequential()
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[0]))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[0]))
        model.add(Dense(100, input_shape=(1,)))
        model.add(Activation(act_func[0]))
        model.add(Dense(1, input_shape=(1,)))
        model.add(Activation(act_func[0]))
        
    models.append(model)

# compile the models choosing optimizer, loss and metrics objects
i=0
for model in models:
    if i<3:
        model.compile(optimizer=opt[0], loss=loss[0], metrics=['mse'])
    if i>=3 & i<5:
        model.compile(optimizer=opt[i-2], loss=loss[0], metrics=['mse'])
    else:
        model.compile(optimizer=opt[0], loss=[i-4], metrics=['mse'])

#No summary because all descriptions are equal
```

```python
# fit the models using training dataset
# report training progress against validation data
history=[]
for model in models:
    history.append(model.fit(x=x_train, y=y_train, 
              batch_size=64, epochs=200,
              shuffle=True, # a good idea is to shuffle input before at each epoch
              validation_data=(x_valid, y_valid)))
```

```python
# return weights
for model in models:
    model.get_weights()
```

```python
i=1
for model in models:
    print("Model ",i)
    # evaluate model
    score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

    # print performance
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print()
    i+=1
```

```python
i=1
for model in models:
    print("Model ",i)
    # evaluate model with the exact curve
    score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

    # print performance
    print('Test loss:', score[0])
    print('Test accuracy:', score[1])
    print()
    i+=1
```

```python
# look into training history
# Plot training & validation loss values

fig, (fig_1, fig_2) = plt.subplots(1, 2, figsize=(18,6))

i=1
plt.figure(figsize=(12,9))
for h in history:
    if i<=4:
        fig_1.plot(h.history['loss'],label="loss "+str(i))
        fig_1.plot(h.history['val_loss'],label="val_loss "+str(i))
    else:
        fig_2.plot(h.history['loss'],label="loss "+str(i))
        fig_2.plot(h.history['val_loss'],label="val_loss "+str(i))
    i+=1

fig_1.legend(loc='best')
fig_1.set_title("Model loss \n Models 1,2,3,4")
fig_1.set_ylabel('Loss')
fig_1.set_xlabel('Epoch')

fig_2.legend(loc='best')
fig_2.set_title("Model loss \n Models 5,6,7")
fig_2.set_ylabel('Loss')
fig_2.set_xlabel('Epoch')
```

```python
fig, (fig_1, fig_2) = plt.subplots(1, 2, figsize=(18,6))

i=1

for model in models:
    y_predicted_1 = model.predict(x_predicted_1)
    y_predicted_2 = model.predict(x_predicted_2)
    fig_1.scatter(x_predicted_1, y_predicted_1,label="Model "+str(i))
    fig_2.scatter(x_predicted_2, y_predicted_2,label="Model "+str(i))
    i+=1

fig_1.plot(x_valid, y_target,label="target",color='y')
fig_1.set_title("Predicted results \n Range $[-1,1]$")
fig_1.legend()


fig_2.plot(x_valid_2, y_target_2,label="target",color='y')
fig_2.set_title("Predicted results \n Range $[1,3]$")
fig_2.legend()
```

### Osservazioni

- Si pu notare che cambiare la funzione di attivazione comporta delle modifiche notevoli sui risultati delle predizioni all'interno del range $[-1,1]$. Come si pu vedere nel grafico dei risultati previsti e nel grafico della *loss function*, il *Modello 2* e il *Modello 3* differiscono vistosamente da tutti gli altri modelli in quanto presentano una diversa funzione di attivazione.


- Il *Modello 4* e il *Modello 5*, cos come il *Modello 6* e il *Modello 7*, invece, mostrano pressoch gli stessi risultati. Infatti si pu notare che gli errori associati ai modelli in questione sono tutti dello stesso ordine di grandezza. In realt, studiando la letteratura, si scopre che in base al modello da descrivere  consigliabile usare un'adeguata *loss function*. Per problemi di regressione, come quello in esame, si usa in genere il *mse* (*mean square error*); per problemi di classificazione, invece, solitamente si consiglia di usare la funzione di *cross entropy*.


- Come analizzato precedentemente, al di fuori del range $[-1,1]$ le previsioni non seguono l'andamento teorico.

### Conclusioni

Dalle osservazioni condotte si pu concludere quello che  stato previsto teoricamente: l'aumento della complessit del modello migliora il risultato della previsione, tuttavia non migliora indefinitamente all'aumentare della complessit. 

Inoltre il modello risulta efficiente finch questo viene utilizzato all'interno del range di studio.
