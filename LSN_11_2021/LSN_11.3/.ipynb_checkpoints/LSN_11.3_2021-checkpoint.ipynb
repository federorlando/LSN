Test of NNs on a data set of labeled data $(x,y)$, generated by drawing samples from the equation

$$
    y_i= f(x_i) + \eta_i,
$$

where $f(x_i)$ is some fixed, possibly unknown function, and $\eta_i$ is a Gaussian, uncorrelate noise variable such that

$$
\langle \eta_i \rangle=0 \\
\langle \eta_i \eta_j \rangle = \delta_{ij} \sigma
$$

##  <span style="color:blue"> Exercise 11.3

$$
f(x)=sin(x^2+y^2)
$$
for $x \in [-3/2,3/2]$.

```python
#import os
#os.environ['KMP_DUPLICATE_LIB_OK']='True'

import numpy as np
import math
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x,y):
    return math.sin(x**2+y**2)

# generate training inputs
N_t = 1750
N_v = int(N_t/10)
np.random.seed(0)
sigma = 0.3 # noise standard deviation

x_train = np.random.uniform(-1.5, 1.5, N_t)
y_train = np.random.uniform(-1.5, 1.5, N_t)
x_valid = np.random.uniform(-1.5, 1.5, N_v)
y_valid = np.random.uniform(-1.5, 1.5, N_v)

xy_train=np.array([
    [x_train[i],y_train[i]] for i in range(N_t)
])
xy_valid=np.array([
    [x_valid[i],y_valid[i]] for i in range(N_v)
])


z_target = ([
    f(x_valid[i],y_valid[i]) # ideal (target) function
    for i in range(N_v)
])
z_train = np.array([
    np.random.normal(f(x_train[i],y_train[i]), sigma)
    for i in range(N_t)
]) 
z_valid = np.array([
    np.random.normal(f(x_valid[i],y_valid[i]), sigma)
    for i in range(N_v)
])

#x_true,y_true and z_true to be used for the 3Dplot comparison
x_true = np.random.uniform(-1.5, 1.5, N_t)
y_true = np.random.uniform(-1.5, 1.5, N_t)
z_true=([
    f(x_true[i],y_true[i]) for i in range(N_t)
])

# plot target dataset
ax = Axes3D(plt.figure())
ax.plot_trisurf(x_true, y_true, z_true, alpha=0.6, color='orange')
ax.scatter(x_valid, y_valid, z_valid, color='blue')
#ax.scatter(x_valid, y_valid, z_target, color='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("$f(x,y)=sin(x^2+y^2)$")
ax.view_init(30, 30)
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(500, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(100, input_shape=(2,)))
model.add(Activation("relu"))
model.add(Dense(1, input_shape=(2,)))
model.add(Activation("relu"))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])

# get a summary of our composed model
model.summary()
```

```python
# fit the model using training dataset
# report training progress against validation data
history = model.fit(x=xy_train, y=z_train,
          batch_size=128, epochs=300,
          shuffle=True, # a good idea is to shuffle input before at each epoch
          validation_data=(xy_valid, z_valid))
```

```python
# evaluate model
score = model.evaluate(xy_train, z_train, batch_size=32, verbose=1)

# print performance
print('VALIDATION AGAINST Y_TRAIN')
print('Loss:', score[0])

# evaluate model
score = model.evaluate(xy_valid, z_valid, batch_size=32, verbose=1)

# print performance
print('VALIDATION AGAINST Y_VALID')
print('Loss:', score[0])
```

```python
# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train ($E_{in}$)', 'Test ($E_{out}$)'], loc='best')
plt.show()
```

```python
x_predicted = np.random.uniform(-1.5, 1.5, 1000)
y_predicted = np.random.uniform(-1.5, 1.5, 1000)
xy_predicted = np.array([
    [x_predicted[i],y_predicted[i]] for i in range(1000)
])
z_predicted = model.predict(xy_predicted)

ax = Axes3D(plt.figure())
ax.plot_trisurf(x_true, y_true, z_true,alpha=0.3,label="true")
ax.scatter(x_predicted, y_predicted, z_predicted, color='r',label="predicted")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title("Predicted results")
#ax.legend()
ax.view_init(15, 35)
```

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.lines as lines

Ndata = 6
absc = np.zeros(Ndata)
testloss = np.zeros(Ndata)
valloss = np.zeros(Ndata)

file = open("./sigma_Ntrain1000_epoch300")
line = file.readlines()
for i in range(Ndata):
    data=line[i].split(" ")
    absc[i]=data[0]
    testloss[i]=data[1]
    valloss[i]=data[2]

plt.figure()

plt.plot(absc,testloss,label='Train')
plt.plot(absc,valloss,label='Validation')
plt.xlabel("$\sigma$")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss")

plt.show()
```

Increasing $\sigma$, loss on both training and validation set increases. Interestingly, for low $\sigma$, the model performs better on the validation set than on the training set (run with $N_{train}=1000$ and 300 epochs).

```python
Ndata = 7
absc = np.zeros(Ndata)
testloss = np.zeros(Ndata)
valloss = np.zeros(Ndata)

file = open("./epoch_sigma0.0_Ntrain1000")
line = file.readlines()
for i in range(Ndata):
    data=line[i].split(" ")
    absc[i]=data[0]
    testloss[i]=data[1]
    valloss[i]=data[2]

plt.figure(figsize=(16, 5))

plt.subplot(1,2,1)
plt.plot(absc,testloss,label='Train')
plt.plot(absc,valloss,label='Validation')
plt.xlabel("epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss ($\sigma$=0.0)")

file = open("./epoch_sigma0.3_Ntrain1000")
line = file.readlines()
for i in range(Ndata):
    data=line[i].split(" ")
    absc[i]=data[0]
    testloss[i]=data[1]
    valloss[i]=data[2]

plt.subplot(1,2,2)
plt.plot(absc,testloss,label='Train')
plt.plot(absc,valloss,label='Validation')
plt.xlabel("epoch")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss ($\sigma$=0.3)")

plt.subplots_adjust(wspace=0.25)
plt.show()
```

Fixing $N_{train}=1000$ and increasing the epochs, losses on both training and validation sets saturate to costant (different) values. Again, we see that in the $\sigma=0$ case the validation curve lies below. In the $\sigma=0.3$ case the opposite holds; also, a glance at the *y* axis scale shows that the saturation values are higher.

```python
Ndata = 8
absc = np.zeros(Ndata)
testloss = np.zeros(Ndata)
valloss = np.zeros(Ndata)

file = open("./Ntrain_sigma0.0_epoch300")
line = file.readlines()
for i in range(Ndata):
    data=line[i].split(" ")
    absc[i]=data[0]
    testloss[i]=data[1]
    valloss[i]=data[2]

plt.figure(figsize=(16, 5))

plt.subplot(1,2,1)
plt.plot(absc,testloss,label='Train')
plt.plot(absc,valloss,label='Validation')
plt.xlabel("N_train")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss ($\sigma$=0.0)")

file = open("./Ntrain_sigma0.3_epoch300")
line = file.readlines()
for i in range(Ndata):
    data=line[i].split(" ")
    absc[i]=data[0]
    testloss[i]=data[1]
    valloss[i]=data[2]

plt.subplot(1,2,2)
plt.plot(absc,testloss,label='Train')
plt.plot(absc,valloss,label='Validation')
plt.xlabel("N_train")
plt.ylabel("Loss")
plt.legend()
plt.title("Loss ($\sigma$=0.3)")

plt.subplots_adjust(wspace=0.25)
plt.show()
```

Fixing $N_{epoch}=300$ and increasing $N_{train}$, one gets a decreasing train with a peculiarity, in that the validation and training curves cross. Choosing here $N_{train}\gtrsim 1250$ seems more suitable. The effect of $\sigma$ is the same as above.
