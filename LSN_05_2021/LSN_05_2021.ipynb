Use the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$, where
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{-3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{-5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$

Use the sampled positions to estimate $\left\langle r \right\rangle_{\Psi_{1,0,0}}$ and $\left\langle r \right\rangle_{\Psi_{2,1,0}}$, with target values

$$
\left\langle r \right\rangle_{\Psi_{1,0,0}} = \frac{3}{2}a_0
\quad
\left\langle r \right\rangle_{\Psi_{2,1,0}} = 5 a_0
$$

Blocking parameters: 10$^6$ time steps, sorted into 100 blocks.

###  Ground state $ \left | 1,0,0 \right \rangle $

```python
import math
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

N = 100
M = 1000000
x = np.zeros(M)
y = np.zeros(M)
z = np.zeros(M)
x2 = np.zeros(M)
y2 = np.zeros(M)
z2 = np.zeros(M)
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
sum_prog2 = np.zeros(N)
err_prog2 = np.zeros(N)
steps = np.arange(N)

#Plot <r> 

file1 = open("./LSN_05.1/output_100_uniform.txt")
line1 = file1.readlines()
for i in range(N):
    data1 = line1[i].split("   ")
    #steps[i] = data1[0]
    sum_prog[i] = data1[1]
    err_prog[i] = data1[2]    
file1.close()

file2 = open("./LSN_05.1/output_100_gauss.txt")
line2 = file2.readlines()
for i in range(N):
    data2 = line2[i].split("   ")
    #steps[i] = data1[0]
    sum_prog2[i] = data2[1]
    err_prog2[i] = data2[2]    
file2.close()

plt.figure(figsize=(16, 5))

plt.subplot(1, 2, 1)
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.xlabel("N° steps")
plt.ylabel("<r> - 1.5 [$a_0$]")
plt.grid(True)
plt.title("<r> - ground state 1s - uniform T(x|y)")

plt.subplot(1, 2, 2)
plt.errorbar(steps, sum_prog2, yerr = err_prog2, color='green')
plt.xlabel("N° steps")
plt.ylabel("<r> - 1.5 [$a_0$]")
plt.ylim(-0.07, 0.05) 
plt.grid(True)
plt.title("<r> - ground state 1s - gaussian T(x|y)")

plt.subplots_adjust(wspace=0.25)
plt.show()
```

```python
#Plot 3D points

file3 = open("./LSN_05.1/output_100_uniform_points.txt")
line3 = file3.readlines()
for i in range(M):
    data3 = line3[i].split("   ")
    x[i] = data3[0]
    y[i] = data3[1]
    z[i] = data3[2]
file3.close()

file4 = open("./LSN_05.1/output_100_gauss_points.txt")
line4 = file4.readlines()
for i in range(M):
    data4 = line4[i].split("   ")
    x2[i] = data4[0]
    y2[i] = data4[1]
    z2[i] = data4[2]
file4.close()
    
fig = plt.figure(figsize=(16, 7))
ax = fig.add_subplot(1, 2, 1, projection='3d')
#ax = Axes3D(fig)
ax.scatter(x, y, z, c=z, marker='.')
ax.set_xlabel('x [$a_0$]')
ax.set_ylabel('y [$a_0$]')
ax.set_zlabel('z [$a_0$]')
ax.set_title('Sampled region - uniform T(x|y)')
ax.view_init(10, 30)

ax = fig.add_subplot(1, 2, 2, projection='3d')
#ax = Axes3D(fig)
ax.scatter(x2, y2, z2, c=z, marker='.')
ax.set_xlabel('x [$a_0$]')
ax.set_ylabel('y [$a_0$]')
ax.set_zlabel('z [$a_0$]')
ax.set_title('Sampled region - gaussian T(x|y)')
ax.view_init(10, 30)

plt.tight_layout()
plt.show()
```

###  Comments
* For both PDFs, the sampled points account for the right symmetry of the orbitals: spherical shape for 1s, dumbbell-like for 2p$_z$
* No substantial difference appears when using a Gaussian transfer matrix instead of a uniform one (as expected)
* The Metropolis algorithm has been implemented in order to obtain a $\simeq$ 50% probability of acceptance. This is obtained via a preliminary Metropolis run. For instance, in the uniform-T case the deviations from the 50% targeted value turned out to be:
    * wfc $\psi_{1,0,0}$: 0.16%
    * wfc $\psi_{2,1,0}$: 0.55%
* All of the above has been done with the origin of $R^3$ as a starting point. This is a reasonable choice, since it is close to the region of $R^3$ where the PDFs are appreciably nonzero (even though $\psi_{2,1,0}(0,0,0)=0$). Conversely, starting very far from the origin leads to a poor sampling. A test has been done for the 1s PDF (with uniform T): with (10,10,10) as starting point the graph still looks fine and the distribution of points starts to deviate slightly from the spherical shape; starting even further away, e.g. (100,100,100), the estimate for $\left\langle r \right\rangle$ turns out to be poor and so is the distribution of points, which looks like a sort of "strip" instead of a sphere.
