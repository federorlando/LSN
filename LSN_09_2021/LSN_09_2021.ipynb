Solution of the TSP with a genetic algorithm (GA), applied to

- 34 cities randomly placed on a circumference
- 34 cities randomly placed inside a square

Results are shown via:

- a picture of the cost function, averaged on the best half of the population, vs generations 
- a picture of the best path in cartesian coordinates

#### *Brief description of the algorithm:*

- Each run consists of $10^4$ generations (but only 2000 are shown in the graphs, since the loss function decay is less appreciable on the $10^4$ scale). The population is set to 46 chromosomes 
- Selection: after evaluation of the current population's fitness, a random number is generated with exponential PDF and summed to the minimum fitness. Then the function searches for (and selects) the chromomome whose fitness is closest to such value.
- Mutation: the four suggested procedures are implemented, each with $P_m =  0.05$. Avoiding touching the first city is either implemented directly in the functions or managed in the main() by careful choice of indices.

### Unit circle

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.lines as lines

N = 2000
avg = np.zeros(N)
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
x = np.arange(N)
N_points = 34
circ_x=np.zeros(N_points)
circ_y=np.zeros(N_points)

file = open("./LSN_09.1/circle_output.txt")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg[i]=data[0]
    sum_prog[i]=data[1]
    err_prog[i]=data[2]

file=open("./LSN_09.1/circle_path.txt")
line=file.readlines()
for i in range(N_points):
    data=line[i].split(" ")
    circ_x[i]=data[0]
    circ_y[i]=data[1]

order=np.zeros(N_points,dtype=int)
for i in range(N_points):
    order[i]=line[N_points+1+i]
file.close()

fig_circ, (ax_circ1, ax_circ2) = plt.subplots(1, 2, figsize=(12, 4))

#ax_circ1.errorbar(x,sum_prog,yerr=err_prog,color='orange')
ax_circ1.plot(x,avg,color='orange')
ax_circ1.set_xlabel("N° generations")
ax_circ1.set_ylabel("<$L^{(1)}$>")
ax_circ1.set_title("Averaged best path <$L^{(1)}$>")
#ax_circ1.set_xscale("log")

ax_circ2.scatter(circ_x,circ_y,color='black')
for i in range(1,N_points):
    ax_circ2.add_line(lines.Line2D([circ_x[order[i-1]-1],circ_x[order[i]-1]],[circ_y[order[i-1]-1],circ_y[order[i]-1]],color='orange'))
ax_circ2.add_line(lines.Line2D([circ_x[order[N_points-1]-1],circ_x[order[0]-1]],[circ_y[order[N_points-1]-1],circ_y[order[0]-1]],color='orange'))
ax_circ2.scatter(circ_x[order[0]-1],circ_y[order[0]-1],marker='*',s=200,color='red',label='start')
ax_circ2.set_xlabel("x")
ax_circ2.set_ylabel("y")
ax_circ2.set_title("Best path")
ax_circ2.legend()

plt.show()
```

The final estimate for the cost in the unit circle case is 6.30923. This is remarkably close to $2\pi$, the best path in this case clearly being one that follows the circle, either CW or CCW.

### Square [0,1] x [0,1]

```python
file=open("./LSN_09.1/square_output.txt")
line=file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg[i]=data[0]
    sum_prog[i]=data[1]
    err_prog[i]=data[2]

file=open("./LSN_09.1/square_path.txt")
line=file.readlines()
for i in range(N_points):
    data=line[i].split(" ")
    circ_x[i]=data[0]
    circ_y[i]=data[1]

for i in range(N_points):
    order[i]=line[N_points+1+i]
file.close()

fig_circ, (ax_circ1, ax_circ2) = plt.subplots(1, 2, figsize=(12, 4))

#ax_circ1.errorbar(x,sum_prog,yerr=err_prog,color='green')
ax_circ1.plot(x,avg,color='green')
ax_circ1.set_xlabel("N° generations")
ax_circ1.set_ylabel("<$L^{(1)}$>")
ax_circ1.set_title("Averaged best path <$L^{(1)}$>")

ax_circ2.scatter(circ_x,circ_y,color='black')
for i in range(1,N_points):
    ax_circ2.add_line(lines.Line2D([circ_x[order[i-1]-1],circ_x[order[i]-1]],[circ_y[order[i-1]-1],circ_y[order[i]-1]],color='g'))
ax_circ2.add_line(lines.Line2D([circ_x[order[N_points-1]-1],circ_x[order[0]-1]],[circ_y[order[N_points-1]-1],circ_y[order[0]-1]],color='g'))
ax_circ2.scatter(circ_x[order[0]-1],circ_y[order[0]-1],marker='*',s=200,color='red',label='start')
ax_circ2.set_xlabel("x")
ax_circ2.set_ylabel("y")
ax_circ2.set_title("Best path")
ax_circ2.legend()

plt.show()
```

The final estimate for the cost in the unit circle case is 4.28763, with the path being now tangled up. Unlike for the unit circle, this minimum value is strongly dependent on the completion couple for the random number generator seed (e.g. it can easily exceed 5).
