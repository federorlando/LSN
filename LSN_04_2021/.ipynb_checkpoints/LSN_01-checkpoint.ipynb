##  <span style="color:blue"> Exercise 1.1

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N = 100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
sum_prog2 = np.zeros(N)
err_prog2 = np.zeros(N)
steps = np.arange(N)

#Plot <r>, <(r-0.5)^2>

file = open("./LSN_01.1/output_1.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split("   ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
    
file2 = open("./LSN_01.1/output_2.txt")
line2 = file2.readlines()
for i in range(N):
    b = line2[i].split("   ")
    sum_prog2[i] = b[1]
    err_prog2[i] = b[2]
    
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.xlabel("N blocks")
plt.ylabel("<r> - 0.5")
plt.grid(True)
plt.title("<r>")

plt.subplot(1, 2, 2)
plt.errorbar(steps, sum_prog2, yerr = err_prog2, color='orange')
plt.xlabel("N blocks")
plt.ylabel("<(r-0.5)^2> - 1/12")
plt.grid(True)
plt.title("$\sigma^2$")

plt.subplots_adjust(wspace=0.25)
plt.show()

#Plot chi_squared

file3 = open("./LSN_01.1/output_3.txt")
line3 = file3.readlines()
for i in range(N):
    c = line3[i].split(" ")
    sum_prog[i] = c[1]
    err_prog[i] = c[2]
    
plt.figure()
plt.errorbar(steps, sum_prog, yerr = err_prog, color='green')
plt.xlabel("N blocks")
plt.ylabel("$\chi^2$")
plt.grid(True)
plt.title("$\chi^2$")
plt.show()
```

##  <span style="color:blue"> Exercise 1.2

```python
M = 10000
s = np.zeros(M)
e = np.zeros(M)
l = np.zeros(M)

#N=1

file = open("./LSN_01.2/output_1.txt")
line = file.readlines()
for i in range(M):
    a = line[i].split("          ")
    s[i] = a[0]
    e[i] = a[1]
    l[i] = a[2]
file.close()

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.hist(s,50,range=(-0.2,1.2))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. U[0,1]')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(e,50,color='orange',range=(0,7))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. Exponential ($\lambda$=1)')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(l,50,color='g',range=(-9,9))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. Lorentzian ($\mu$=0, $\Gamma$=1)')
plt.grid(True)

plt.subplots_adjust(wspace=0.25)
plt.show()
```

##  <span style="color:blue"> Exercise 1.3

```python
N = 100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
steps = np.arange(N)

file = open("./LSN_01.3/output.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split(" ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
    
plt.figure()
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.axhline(y=math.pi, color='orange', label='y=$\pi$')
plt.xlabel("N blocks")
plt.ylabel("Estimated $\pi$")
plt.ylim(3.12, 3.18) 
plt.grid(True)
plt.title("$\pi$")

plt.show()
```
