##  <span style="color:blue"> Exercise 1.1

1. Estimation of $\langle r \rangle = \int_0^1 r dr = 1/2$ ($M = 10^5$ throws, $N = 10^2$ blocks) 
2. Estimation of $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ ($M = 10^5$ throws, $N = 10^2$ blocks) 
3. Implementation of $\chi^2$ test (expected value $\chi^2 \simeq 100$)

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt

N = 100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
sum_prog2 = np.zeros(N)
err_prog2 = np.zeros(N)
steps = np.arange(N)

#Plot <r>, <(r-0.5)^2>

file = open("./LSN_01.1/output_1.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split("   ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
    
file2 = open("./LSN_01.1/output_2.txt")
line2 = file2.readlines()
for i in range(N):
    b = line2[i].split("   ")
    sum_prog2[i] = b[1]
    err_prog2[i] = b[2]
    
plt.figure(figsize=(15, 4))

plt.subplot(1, 2, 1)
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.xlabel("N° blocks")
plt.ylabel("<r> - 0.5")
plt.grid(True)
plt.title("<r>")

plt.subplot(1, 2, 2)
plt.errorbar(steps, sum_prog2, yerr = err_prog2, color='orange')
plt.xlabel("N° blocks")
plt.ylabel("<(r-0.5)^2> - 1/12")
plt.grid(True)
plt.title("$\sigma^2$")

plt.subplots_adjust(wspace=0.25)
plt.show()

#Plot chi_squared

file3 = open("./LSN_01.1/output_3.txt")
line3 = file3.readlines()
for i in range(N):
    c = line3[i].split(" ")
    sum_prog[i] = c[1]
    err_prog[i] = c[2]
    
plt.figure()
plt.errorbar(steps, sum_prog, yerr = err_prog, color='green')
plt.xlabel("N° blocks")
plt.ylabel("$\chi^2$")
plt.grid(True)
plt.title("$\chi^2$")

plt.show()
```

##  <span style="color:blue"> Exercise 1.2

Plotted are histograms obtained with $10^4$ realizations of $S_N = \frac{1}{N}\sum_{i=1}^N x_i$ (for $N=1, 2, 10, 100$), being $x_i$ a random variable sampled via:

Left: standard dice 

Centre: exponential dice $p(x) = \lambda \exp(-\lambda x)$ (here $\lambda=1$) 

Right: Lorentzian dice $p(x)=\frac{1}{\pi}\frac{\Gamma}{(x-\mu)^2+\Gamma^2}$ (here $\mu=0$, $\Gamma=1$).

```python
M = 10000
s = np.zeros(M)
e = np.zeros(M)
l = np.zeros(M)

#N=1

file = open("./LSN_01.2/output_1.txt")
line = file.readlines()
for i in range(M):
    a = line[i].split("          ")
    s[i] = a[0]
    e[i] = a[1]
    l[i] = a[2]
file.close()

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.hist(s,50,range=(-0.2,1.2))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. U[0,1]')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(e,50,color='orange',range=(0,7))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. Exponential ($\lambda$=1)')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(l,50,color='g',range=(-9,9))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=1. Lorentzian ($\mu$=0, $\Gamma$=1)')
plt.grid(True)

plt.subplots_adjust(wspace=0.25)
plt.show()
```

```python
#N=2
    
file = open("./LSN_01.2/output_2.txt")
line = file.readlines()
for i in range(M):
    a = line[i].split("          ")
    s[i] = a[0]
    e[i] = a[1]
    l[i] = a[2]
file.close()

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.hist(s,50,range=(-0.2,1.2))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=2. U[0,1]')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(e,50,color='orange',range=(0,7))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=2. Exponential ($\lambda$=1)')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(l,50,color='g',range=(-9,9))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=2. Lorentzian ($\mu$=0, $\Gamma$=1)')
plt.grid(True)

plt.subplots_adjust(wspace=0.25)
plt.show()
```

```python
#N=10
    
file = open("./LSN_01.2/output_10.txt")
line = file.readlines()
for i in range(M):
    a = line[i].split("          ")
    s[i] = a[0]
    e[i] = a[1]
    l[i] = a[2]
file.close()

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.hist(s,50,range=(0,1))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=10. U[0,1]')
plt.grid(True)

plt.subplot(1, 3, 2)
plt.hist(e,50,color='orange',range=(0,3))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=10. Exponential ($\lambda$=1)')
plt.grid(True)

plt.subplot(1, 3, 3)
plt.hist(l,50,color='g',range=(-9,9))
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=10. Lorentzian ($\mu$=0, $\Gamma$=1)')
plt.grid(True)

plt.subplots_adjust(wspace=0.25)
plt.show()
```

```python
#N=100

from scipy.optimize import curve_fit

def gauss(x,a,b,c):
    return a*np.exp(-b*(x-c)**2)

def lorentz(x,a,b,c):
    return a/(b**2+(x-c)**2)
    
file = open("./LSN_01.2/output_100.txt")
line = file.readlines()
for i in range(M):
    a = line[i].split("          ")
    s[i] = a[0]
    e[i] = a[1]
    l[i] = a[2]
file.close()

plt.figure(figsize=(16, 4))

plt.subplot(1, 3, 1)
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=100. U[0,1]')
plt.grid(True)
s_heights, s_borders,_ =plt.hist(s,50,range=(0.4,0.6))
s_centers=s_borders[:-1]+np.diff(s_borders)/2
s_int= np.linspace(s_borders[0],s_borders[-1],10000)
s_opt,_= curve_fit(gauss,s_centers,s_heights)
plt.plot(s_int,gauss(s_int,s_opt[0],s_opt[1],s_opt[2]),label='fit',color='r')
plt.legend()
print("Fit parameters for standard dice [a,b,c] =")
print(s_opt)

plt.subplot(1, 3, 2)
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=100. Exponential ($\lambda$=1)')
plt.grid(True)
e_heights, e_borders,_ =plt.hist(e,50,color='orange',range=(0.5,1.5))
e_centers=e_borders[:-1]+np.diff(e_borders)/2
e_int= np.linspace(e_borders[0],e_borders[-1],10000)
e_opt,_= curve_fit(gauss,e_centers,e_heights)
plt.plot(e_int,gauss(e_int,e_opt[0],e_opt[1],e_opt[2]),label='fit',color='r')
plt.legend()
print("Fit parameters for exponential dice [a,b,c] =")
print(e_opt)

plt.subplot(1, 3, 3)
plt.xlabel('Values')
plt.ylabel('Occurrencies')
plt.title('N=100. Lorentzian ($\mu$=0, $\Gamma$=1)')
plt.grid(True)
l_heights, l_borders,_ =plt.hist(l,50,color='g',range=(-9,9))
l_centers=l_borders[:-1]+np.diff(l_borders)/2
l_int= np.linspace(l_borders[0],l_borders[-1],10000)
l_opt,_= curve_fit(lorentz,l_centers,l_heights)
plt.plot(l_int,lorentz(l_int,l_opt[0],l_opt[1],l_opt[2]),label='fit',color='r')
plt.legend()
print("Fit parameters for Lorentzian dice [a,b,c] =")
print(l_opt)

plt.subplots_adjust(wspace=0.25)
plt.show()

print('CLT is verified for the standard and exponential dice. \nConversely, Lorentzian dice belongs to another basin of attraction.')
```

##  <span style="color:blue"> Exercise 1.3

Simulation of Buffon’s experiment: $\pi = \frac{2L}{Pd} = \lim_{N_{thr} \to \infty}\frac{2LN_{thr}}{N_{hit}d}$ (with $d > L$)

```python
N = 100
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
steps = np.arange(N)

file = open("./LSN_01.3/output.txt")
line = file.readlines()
for i in range(N):
    a = line[i].split(" ")
    sum_prog[i] = a[1]
    err_prog[i] = a[2]
file.close()
    
plt.figure()
plt.errorbar(steps, sum_prog, yerr = err_prog)
plt.axhline(y=math.pi, color='orange', label='y=$\pi$')
plt.xlabel("N° blocks")
plt.ylabel("Estimated $\pi$")
plt.ylim(3.13, 3.17) 
plt.grid(True)
plt.title("$\pi$")

plt.show()
```
