Solution of the TSP applied to the 50 American capitals. The code builds on that used in Lecture 9, parallelized with MPI libraries in order to solve the TSP by performing (up to) 10 parallel GA searches of the optimal path. Each node performs an independent GA search, the so-called *continent*, but every $N_{migr}=500$ generations the continents randomly exchange their best individuals.

### Serial

```python
import numpy as np
import math
import matplotlib 
import matplotlib.pyplot as plt
import matplotlib.lines as lines
from matplotlib.figure import Figure
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

N = 2000
avg = np.zeros(N)
sum_prog = np.zeros(N)
err_prog = np.zeros(N)
x = np.arange(N)
N_points = 50
circ_x=np.zeros(N_points)
circ_y=np.zeros(N_points)

file = open("./LSN_10.2_serial/square_output.txt")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg[i]=data[0]
    sum_prog[i]=data[1]
    err_prog[i]=data[2]

file=open("./LSN_10.2_serial/square_path.txt")
line=file.readlines()
for i in range(N_points):
    data=line[i].split(" ")
    circ_x[i]=data[0]
    circ_y[i]=data[1]

order=np.zeros(N_points,dtype=int)
for i in range(N_points):
    order[i]=line[N_points+1+i]
file.close()

fig_circ, (ax_circ1, ax_circ2) = plt.subplots(1, 2, figsize=(12, 4))

#ax_circ1.errorbar(x,sum_prog,yerr=err_prog,color='red')
ax_circ1.plot(x,avg,color='red')
ax_circ1.set_xlabel("N° generations")
ax_circ1.set_ylabel("<$L^{(1)}$>")
#ax_circ1.set_xscale("log")
ax_circ1.set_title("Averaged best path <$L^{(1)}$>")
#ax_circ1.legend()

ax_circ2.scatter(circ_x,circ_y,color='black')
for i in range(1,N_points):
    ax_circ2.add_line(lines.Line2D([circ_x[order[i-1]-1],circ_x[order[i]-1]],[circ_y[order[i-1]-1],circ_y[order[i]-1]],color='red'))
ax_circ2.add_line(lines.Line2D([circ_x[order[N_points-1]-1],circ_x[order[0]-1]],[circ_y[order[N_points-1]-1],circ_y[order[0]-1]],color='red'))
ax_circ2.scatter(circ_x[order[0]-1],circ_y[order[0]-1],marker='*',s=200,color='cyan',label='start')
ax_circ2.set_xlabel("Longitude [° W]")
ax_circ2.set_ylabel("Latitude [° N]")
ax_circ2.set_title("Best path")
ax_circ2.legend()

arr_img = plt.imread("./merlino.jpg")
im = OffsetImage(arr_img, zoom=.065)
ab = AnnotationBbox(im, (0, 0), xycoords='axes fraction',box_alignment=(-1.1,-0.3))
ax_circ2.add_artist(ab)

plt.show()
```

Running serial gives a minimum cost of 276.058.

```python
avg2 = np.zeros(N)
avg4 = np.zeros(N)
avg6 = np.zeros(N)
avg8 = np.zeros(N)
avg10 = np.zeros(N)

proc = np.zeros(6)
best = np.zeros(6)
best_seq = np.zeros(6)

file = open("./LSN_10.2_serial/square_output.txt")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/2nodes/cost_rank1")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg2[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/4nodes/cost_rank1")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg4[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/6nodes/cost_rank1")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg6[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/8nodes/cost_rank1")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg8[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/10nodes/cost_rank1")
line = file.readlines()
for i in range(N):
    data=line[i].split(" ")
    avg10[i]=data[0]
file.close()

file = open("./LSN_10.2_parallel/bestloss_vs_np.txt")
line = file.readlines()
for i in range(6):
    data=line[i].split(" ")
    proc[i]=data[0]
    best[i]=data[1]
    best_seq[i]=data[2]
file.close()

fig_circ, (ax_circ1, ax_circ2) = plt.subplots(1, 2, figsize=(12, 4))

ax_circ1.plot(x,avg,color='red',label='serial')
ax_circ1.plot(x,avg2,color='green',label='np 2')
ax_circ1.plot(x,avg4,color='orange',label='np 4')
ax_circ1.plot(x,avg6,color='blue',label='np 6')
ax_circ1.plot(x,avg8,color='magenta',label='np 8')
ax_circ1.plot(x,avg10,color='cyan',label='np 10')
ax_circ1.set_xlabel("N° generations")
ax_circ1.set_ylabel("<$L^{(1)}$>")
ax_circ1.set_xscale("log")
ax_circ1.set_title("Averaged best path <$L^{(1)}$>")
ax_circ1.legend()

ax_circ2.plot(proc,best,color='green',label='random')
ax_circ2.plot(proc,best_seq,color='orange',label='sequential')
ax_circ2.set_xlabel("N° nodes")
ax_circ2.set_ylabel("$L^{(1)}$")
ax_circ2.set_title("Best path $L^{(1)}$ vs np")
ax_circ2.legend()

plt.show()
```

* Coupling nodes sequentially (1-2, 3-4...): already doubling the number of nodes, we get a considerable improvement, which becomes smaller (or even vanishes) as the number of nodes is further increased.
* Coupling nodes randomly: improvement is even stronger with np=2 but from then on the plot saturates (by the way, to a value slightly larger than in the sequential case). The graph on the left, only showing two curves, refers to this case.

As noted in Lecture 9, these decreasing trends are explained by the non-negligible dependence of the GA search on the seed completion couple in *Primes*. Exploring further and further in that file may - and indeed does - allow us to stumble into couples that give a lower outcome for $L^{(1)}$.
